```python
import pandas as pd
```

```python
adversary = pd.read_csv('adversary 2.csv')
friend = pd.read_csv('friend.csv')
code = pd.read_csv('COW country codes.csv')
year_0 = 1996
```

```python
ccode = code['CCode']
state = code['StateAbb']
name = code['StateNme']
code_dic = {}
code2name= {} 
for (i,u) in enumerate(zip(ccode,state,name)):
    code_dic[u[1]] = u[0]
    code2name[u[0]] = u[2]
print(code_dic)
```

```python
#read coordinate
lng_lat = pd.read_csv('坐标.csv')
c = lng_lat['Country']
lat = lng_lat['Latitude']
lng = lng_lat['Longitude']
c_geo = {}
for (i,u) in enumerate(zip(c,lat,lng)):
    print(u)
    lat_0 = u[1][0:-2]
    a,b = lat_0.split('°')
    lat_0 = float(a)+float(b)/60
    if 'S' in u[1]:
        lat_0 = -lat_0
        
    lng_0 = u[2][0:-2]
    a,b = lng_0.split('°')
    lng_0 = float(a)+float(b)/60
    if 'W' in u[2]:
        lng_0 = -lng_0
    c_geo[u[0]] = (lat_0,lng_0)
# print(c_geo)

power_csv= pd.read_csv('NMC-60-wsupplementary.csv')
year = power_csv['year']
power = power_csv['cinc']
cc = power_csv['ccode']

c_p = {}
for (i,u) in enumerate(zip(cc,year,power)):
    if int(u[1])==year_0:
        c_p[u[0]] = u[2]
print(c_p)
```

```python
# read the relation csv 
countries = set()
fri = []
adv = []
a = adversary['statea']
b = adversary['stateb']
y = adversary['year']
for (i,u) in enumerate(zip(a,b,y)):
    if int(u[2])==year_0:
        if code2name[u[1]] not in c_geo or (code2name[u[0]] not in c_geo) or (u[1] not in c_p.keys()) or (u[0] not in c_p.keys()):
            continue
        countries.add(u[0])
        countries.add(u[1])
        adv.append(u)

friend = friend.fillna('')
a = friend['ccode1']
b = friend['ccode2']
y1 = friend['dyad_st_year']
y2 = friend['dyad_end_year']

for (i,u) in enumerate(zip(a,b,y1,y2)):
#     print(u)
    if int(u[2])<year_0 and (u[3]=='' or int(u[3])>year_0): 
        if (code2name[u[1]] not in c_geo) or (code2name[u[0]] not in c_geo) or (u[1] not in c_p.keys()) or (u[0] not in c_p.keys()):
            continue
        countries.add(u[0])
        countries.add(u[1])
        fri.append(u)
# print(fri)
# print(c_geo)
# print(c_p)
for u in c_p:
#     print(code2name[u])
    if code2name[u] in c_geo:
        countries.add(u)
# print(c_geo)
# print(code2name[710])
# print(countries)
```

```python
# establish index for countries
i = 0
n_code = {}
n_code2ccode = {}
relation = set()
# print(countries)
for c in countries:
    n_code[c] = i
    n_code2ccode[i] = c
    i+=1
# print(n_code2ccode)
# print(n_code)
for u in fri:
    relation.add((n_code[u[0]],n_code[u[1]],1,1))

for u in adv:
    relation.add((n_code[u[0]],n_code[u[1]],-1,1))
#print(relation)
```

```python
import folium
m = folium.Map()



for r in fri:
    c1 = code2name[r[0]]
    c2 = code2name[r[1]]
    if c1 not in c_geo or c2 not in c_geo:
        continue
    folium.PolyLine(
    locations=[
        [c_geo[c1][0],c_geo[c1][1]],
        [c_geo[c2][0],c_geo[c2][1]],
    ],
    opacity=0.3,
    weight=1,
    color='green'
).add_to(m)
    

for r in adv:
    c1 = code2name[r[0]]
    c2 = code2name[r[1]]
    if c1 not in c_geo or c2 not in c_geo:
        continue
    folium.PolyLine(
    locations=[
        [c_geo[c1][0],c_geo[c1][1]],
        [c_geo[c2][0],c_geo[c2][1]],
    ],
    opacity=0.3,
    weight=1,
#     edge_width 
    color='red'
).add_to(m)

# for code in countries:
#     folium.Marker([c_geo[code2name[code]][0],c_geo[code2name[code]][1]], popup=str(code)).add_to(m)
for code in countries:
    name = code2name[code]
    if name not in c_geo:
        continue
    lat,lng = c_geo[name]
#     print(lat,lng)
    folium.CircleMarker(
    location=[lat,lng],
    radius=2,
    fill=True,
    fill_color='blue',
    color='blue',
    weight=2,
    opacity =0.3
    
    ).add_to(m)
    
    m.save(str(year_0)+'.html')

    m
```
